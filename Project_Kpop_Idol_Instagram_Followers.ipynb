# Exploring Data

```python
# กรองข้อมูลไอดอลที่มีผู้ติดตามตั้งแต่ 5000000 ถึง 100000000
idol_followers = idol[(idol['followers'] >= 5000000) & (idol['followers']<= 100000000)]
idol_followers
```

```python
csv_file_path = '/content/drive/MyDrive/DataClean/idols_data_Clean_followersUp5000000.csv'  # ปรับเปลี่ยนตามต้องการ

# เขียน DataFrame ลงในไฟล์ CSV
idol_followers.to_csv(csv_file_path, index=False)

print(f'ไฟล์ CSV ถูกสร้างเรียบร้อย: {csv_file_path}')
```

```python
# ค่าเฉลี่ยของผู้ติดตามของไอดอลในแต่ละค่ายสังกัด
idol_company_mean = idol_followers.groupby('company')['followers'].mean().reset_index()
idol_company_mean
```

```python
# ค่าเฉลี่ยของผู้ติดตามของไอดอลในแต่ละค่ายสังกัด 5 อันดับ จากมากไปน้อย
idol_company_mean_top5 = idol_company_mean.sort_values(by='followers', ascending=False).head(5)
idol_company_mean_top5
```

```python
# ยอดผู้ติดตามของไอดอล 5 อันดับ จากมากไปน้อย
idol_followers_top5 = idol_followers.sort_values(by='followers', ascending=False).head(5)
idol_followers_top5
```

```python
# จำนวนไอดอลที่มีจากประเทศต่าง ๆ
idol_popular_country = idol_followers['country'].value_counts()
idol_popular_country
```

```python
# กรองข้อมูลตามชื่อ Group ของไอดอลและเรียงยอดผู้ติดตามจากมากไปน้อย
group = (idol_followers['group'] == 'LE SSERAFIM')
idol_followers[group].sort_values(by='followers', ascending=False)
```
